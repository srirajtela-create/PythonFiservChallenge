**Exercise 10-1** (scores_by_score.py) Redo scores.py, printing out the students in descending order by score. TIP You will not need to change anything in scores.py other than the loop that prints out the names and scores.

```python
def get_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 85:
        return 'B'
    elif score >= 80:
        return 'C'
    elif score >= 70:
        return 'D'
    else:
        return 'F'

scores = {}

file_path = "/scores.txt"
with open(file_path, 'r') as f:
    for line in f:
        parts = line.strip().split()
        if len(parts) == 2:
            name, score = parts
            scores[name] = int(score)

print("Name\tScore\tGrade")

for name, score in sorted(scores.items(), key=lambda item: item[1]):
    grade = get_grade(score)
    print(f"{name}\t{score}\t{grade}")

avg_score = sum(scores.values()) / len(scores)
print(f"\nAverage Score: {avg_score:.2f}")
```

**Exercise 10-2** (alt_sorted.py) Read in the file alt.txt. Put all the words that start with 'a' in to a file named a_sorted.txt, in sorted order. Put all the words that start with 'b' in b_sorted.txt, in reverse sorted order. TIP Read through the file once, putting lines into two lists

```python
with open("/alt1.txt", "r") as infile:
    words = infile.read().split()  # read all words (split by whitespace)

a_words = []
b_words = []

# Separate words starting with 'a' or 'b'
for word in words:
    if word.lower().startswith('a'):
        a_words.append(word)
    elif word.lower().startswith('b'):
        b_words.append(word)

# Sort the lists
a_words.sort(reverse=True)  # reverse order (Z → A)
b_words.sort(reverse=True)  # reverse order (Z → A)

# Write 'a' words
with open("awords.txt", "w") as a_file:
    for word in a_words:
        a_file.write(word + "\n")

# Write 'b' words
with open("bwords.txt", "w") as b_file:
    for word in b_words:
        b_file.write(word + "\n")

print("Files created: awords.txt and bwords.txt")
```

**Exercise 10-3** (sort_fruit.py) Using the file fruit.txt, print it out: • sorted by name case-sensitively (the default) • sorted by name case-insensitively (ignoring case) • sorted by length of name, then by name • sorted by the 2nd letter of the name, then the first letter

```python
with open("/fruits.txt", "r") as f:
    fruits = [line.strip() for line in f if line.strip()]  # remove empty lines

# Case-sensitive sort (default)
print("1. Sorted by name (case-sensitive):")
for fruit in sorted(fruits):
    print(fruit)

# Case-insensitive sort
print("\n2. Sorted by name (case-insensitive):")
for fruit in sorted(fruits, key=str.lower):
    print(fruit)

#Sorted by length, then by name
print("\n3. Sorted by length of name, then by name:")
for fruit in sorted(fruits, key=lambda x: (len(x), x)):
    print(fruit)

#Sorted by 2nd letter, then by 1st letter
print("\n4. Sorted by 2nd letter, then by 1st letter:")
for fruit in sorted(fruits, key=lambda x: (x[1] if len(x) > 1 else '', x[0])):
    print(fruit)
```

**Exercise 10-4** (sort_presidents.py) Using the file presidents.txt, print out the presidents' first name, last name, and state of birth, sorted by last name, then first name. TIP Use the split() method on each line to get the individual field

```python
# sort_presidents.py

# Read file and split lines
presidents = []
with open("/presidentslist.txt", "r") as f:
    for line in f:
        parts = line.strip().split()
        if len(parts) == 3:
            presidents.append(parts)
        else:
            print(f"Skipping line with unexpected format: {line.strip()}")


# Sort by last name, then first name
sorted_presidents = sorted(presidents, key=lambda x: (x[1], x[0]))

# Print header
print(f"{'First Name':<15}{'Last Name':<20}{'State of Birth':<20}")
print("-" * 55)

# Print each row with alignment
for first, last, state in sorted_presidents:
    print(f"{first:<20}{last:<25}{state:<20}")
```

**Exercise 13-2** (mark_big_words.py) Copy parrot.txt to bigwords.txt adding asterisks around all words that are 8 or more characters long. HINT: Use the \b anchor to indicate beginning or end of a word

```python
import re

input_file = "/Parrot.txt"
output_file = "bigwords.txt"

pattern = r'\b(\w{8,})\b'

with open(input_file, 'r') as infile, open(output_file, 'w') as outfile:
    for line in infile:
        new_line = re.sub(pattern, r'*\1*', line)
        outfile.write(new_line)

print(f"Processed file saved as {output_file}")
```

**Exercise 13-4** (word_freq.py) Write a script that will read a text file and print out a list of all the words in the file, normalized to lower case, and with the number of times that word occurred in the file. Use the regular expression [^\w']+ for splitting each line into words. Test with any of the text files in the DATA folder. HINT: Use a dictionary for counting. NOTE The specified pattern matches one or more characters that are neither letters, digits, underscores, nor apostrophes

```python
import re

filename = input("Enter filename: ")

word_counts = {}

split_pattern = r"[^\w']+"

try:
    with open(filename, 'r') as f:
        for line in f:
            words = re.split(split_pattern, line.strip().lower())
            for word in words:
                if word:
                    word_counts[word] = word_counts.get(word, 0) + 1

    print("\nWord Frequency List:\n")
    for word in sorted(word_counts):
        print(f"{word:<15} {word_counts[word]}")

except FileNotFoundError:
    print("Error: File not found. Please check the filename and path.")
```

**Exercise 14-1** (knight.py, knight_info.py) Part 1: Create a module which defines a class named Knight. The initializer for the class should expect the knight’s name as a parameter. Get the data from knights.txt to initialize the object. The object should have the following read-only properties: name title favorite_color quest comment Example from knight import Knight k = Knight("Arthur") print k.favorite_color

```python
import os

# ----- Define the Knight class -----
class Knight:
    def __init__(self, name, data_file="/knights (1).txt"):
        self._name = name
        self._title = None
        self._favorite_color = None
        self._quest = None
        self._comment = None

        # Check if the data file exists
        if not os.path.exists(data_file):
            raise FileNotFoundError(f"{data_file} not found.")

        # Read the knight data from file
        with open(data_file, "r") as f:
            for line in f:
                parts = [p.strip() for p in line.strip().split(",")]
                if len(parts) != 5:
                    continue  # skip invalid lines
                n, t, color, quest, comment = parts
                if n.lower() == name.lower():
                    self._title = t
                    self._favorite_color = color
                    self._quest = quest
                    self._comment = comment
                    break

        # If knight not found
        if self._title is None:
            raise ValueError(f"No knight found with name '{name}'")

    # Read-only properties
    @property
    def name(self):
        return self._name

    @property
    def title(self):
        return self._title

    @property
    def favorite_color(self):
        return self._favorite_color

    @property
    def quest(self):
        return self._quest

    @property
    def comment(self):
        return self._comment

if __name__ == "__main__":
    try:
        k = Knight("Arthur")
        print("Name:", k.name)
        print("Title:", k.title)
        print("Favorite Color:", k.favorite_color)
        print("Quest:", k.quest)
        print("Comment:", k.comment)
    except Exception as e:
        print(e)
```
